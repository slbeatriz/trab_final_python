```python
import pandas as pd
import matplotlib.pyplot as plt

tabela = pd.read_csv('dados_cursos.csv')

dados_grafico = tabela['GRAU'].value_counts()

dados_grafico.plot.pie(autopct='%1.1f%%', startangle=30, figsize=(6, 6))
plt.title('Matrículas de Acordo com Grau')

plt.savefig('grau.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

tabela = pd.read_csv('dados_cursos.csv')

dados_grafico = tabela['CATEGORIA_ADMINISTRATIVA'].value_counts()

explode = [0.1 + i * 0.1 for i in range(len(dados_grafico))]

dados_grafico.plot.pie(autopct='%1.1f%%', startangle=-20, figsize=(8, 8), explode=explode)

plt.title('Distribuição de Instituições de Ensino Superior')

plt.legend(dados_grafico.index, loc="best")

plt.savefig('intituicoes.png', bbox_inches='tight')
plt.show()
```
